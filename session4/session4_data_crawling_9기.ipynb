### Requests

```python
import numpy as np
import pandas as pd
import requests # urllib 말고 requests 써주세요 (urllib은 구식입니다!)
```

```python
# requests.post에 url을 전달해주면, 해당 url의 html을 가져와줍니다. 또는 requests.get으로 불러오셔도 좋습니다.
# get은 데이터를 쿼리스트링에 담아 전송 -> 서버의 정보를 조회만 할 수 있음
# post는 데이터를 body에 담아 전송 -> 길이 제한 없이 대용량 데이터 전송 가능, 리소스를 생성 및 변경, 정보를 보낼 때 눈에 보이지 않도록 보낼 수 있음

url = "https://ghsnu.com/"

gh = requests.get(url) #or gh = requests.post(url)

status = gh.status_code

print("http status code : %s" % status) #200 : 정상, # 400  : Bad Request , # 500 Internal Server Error
```

### BeautifulSoup

```python
from bs4 import BeautifulSoup as bs
import lxml
```

```python
# bs 로 html parsing 하기 !
html =  """<html> <head> </head> <body> <p>hello gh</p> </body></html>"""

# html parser 보다 lxml이 더 빠릅니다.
soup = bs(html, "lxml")

# 원하는 부분 선택해서 출력 !

print(soup.p)

print(soup.text)
```

```python
daily_boxoffice_html = driver.page_source
print(daily_boxoffice_html)

#selenium으로 html을 가져와서 bs로 parsing해보겠습니다. driver자체에서도 find할 수 있는데, 오늘은 일단 parsing으로 하는 것으로 해볼게요.
```

```python
parsed_daily_boxoffice_html = bs(daily_boxoffice_html, 'lxml')  #selenium으로 얻어온 페이지의 source를 bs로 parsing해줍니다!
```

```python
daily_boxoffice = parsed_daily_boxoffice_html.find("div", {"id":"dailyBoxOffice"}) # parsed_daily_boxoffice_html.select("#dailyBoxOffice")[0]과 같습니다!

print(daily_boxoffice.prettify())
```

```python
driver = webdriver.Chrome(path)

driver.get("http://www.kobis.or.kr/kobis/business/stat/boxs/findDailyBoxOfficeList.do")


for i in driver.find_elements_by_class_name("more"):

    for _ in range(4):

        element = i.find_element_by_tag_name("a")
        element.click()


daily_boxoffice_html = driver.page_source
```

```python
parsed_daily_boxoffice_html = bs(daily_boxoffice_html, 'lxml')
result = []

for i in parsed_daily_boxoffice_html.find_all("tbody")[:7]:
    daily_boxoffice = i
    movie_list = daily_boxoffice.find_all("tr")
    result.append(movie_list)
```

```python
# 저번시간에 배운 dataframe 을 사용해 보자!
import pandas as pd

daily_movies = {'영화명' : [], "날짜" : [],
'순위 변동' : [],
'개봉일' : [],
'매출액' : [],
'매출액점유율' : [],
'관객수' : [],
'누적관객수' : [],
'스크린수' : [],
'상영횟수' : []}

for i in range(7):
    for movie in result[i]:
        title_rank = movie.find("td").find_next_sibling("td").text
        daily_movies["날짜"].append(parsed_daily_boxoffice_html.find_all("div", {"class":"board_tit"})[i].text.strip())
        daily_movies['영화명'].append(title_rank[:-5].strip())
        daily_movies['순위 변동'].append(title_rank[-4:].strip())
        daily_movies['개봉일'].append(movie.find("td").find_next_sibling("td").find_next_sibling("td").text.strip())
        daily_movies['매출액'].append(movie.find("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").text.strip())
        daily_movies['매출액점유율'].append(movie.find("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").text.strip())
        daily_movies['관객수'].append(movie.find("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").text.strip())
        daily_movies['누적관객수'].append(movie.find("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").text.strip())
        daily_movies['스크린수'].append(movie.find("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").text.strip())
        daily_movies['상영횟수'].append(movie.find("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").text.strip())
```

### 구글 place api를 사용해보도록 합시다!!

```python
import pandas as pd
import requests
import json
from Key import key # API key는 여러분의 비밀번호와도 같은 거라... 꼭 파일 안에 넣지 말고, 따로 보관을 해주세요. 누군가 여러분의 키를 알고 악용한다면 여러분도 모르게 통장에서 돈이 빠져나갈 수 있습니다...

address_ori = '13201 Ranch Road, 620 North' # 검색하시고 싶은 주소를 적어주세요.
address = address_ori.replace(' ', '%20') # URL사이에 빈공간이 없도록 해주세요. space는 %20으로 대체해줍니다!

fields = 'formatted_address,name,rating,opening_hours,permanently_closed,geometry' 
# API를 요청해서 받고 싶은 정보들을 string으로 적어주세요. API가 제공할 수 있는 fields들은 아래 링크에 정리되어 있습니다.
# https://developers.google.com/maps/documentation/places/web-service/place-data-fields

url = f"https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={address}&inputtype=textquery&fields={fields}&key={key}" # f스트링을 사용하면 보기좋게 string을 정리할 수 있습니다. (파이썬 3.6이상)

response = requests.get(url).json() # API에 요청해볼게요! json()은 json을 dict로 바꿔줍니다.

print(response,'\n') # response를 프린트해봤습니다.
print(response['candidates'][0]['geometry']['location'])
```

```python
address_list = ['서울대 경영대학', '서울대입구역','서울시청'] #리스트에 검색할 장소들을 넣어줍니다!
placesAPI_data = pd.DataFrame(columns = ['formatted_address','name','permanently_closed','geometry','rating']) #데이터프레임에 넣어서 정리를 해 볼 거예요.
fields = 'formatted_address,name,rating,opening_hours,permanently_closed,geometry' 

for i in range(len(address_list)):
    address = address_list[i].replace(' ', '%20') 
    url = f"https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={address}&inputtype=textquery&fields={fields}&key={key}" 
    response = requests.get(url).json()
    placesAPI_data = pd.concat([placesAPI_data, pd.DataFrame(response['candidates'])], ignore_index=True, sort=False) #데이터프레임에 한 행씩 붙여줍니다.
    print(i, " " , response) 
```
