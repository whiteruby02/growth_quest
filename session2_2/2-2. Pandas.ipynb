## Null/NaN - isna, fillna, dropna

```python
# isna : 결측치 존재 확인
# 결측치 -> True
# *.isna()

df.isna()
```

## groupby

```python
# 한 칼럼으로 groupby
# df.groupby('col')

# groupby 단독으로는 활용 불가
print(df.groupby('Pclass'))
```

```python
# Pclass로 groupby 후 각 column별 mean

df.groupby('Pclass').mean()
```

```python
# groupby 한 column의 value가 index로

df.groupby('Pclass').mean().index
```

```python
# 여러 칼럼으로 groupby
# df.groupby(['col1', 'col2',....])......to_frame()
# *.to_frame() 없으면 DataFrame 형태로 출력 안 됨

# Pclass, Sex로 groupby 후 Age 칼럼에 대해 sum -> DataFrame으로
df.groupby(['Pclass', 'Sex'])['Age'].sum().to_frame()
```

```python
# 여러 칼럼으로 groupby -> multi-index

df.groupby(['Pclass', 'Sex'])['Age'].sum().to_frame().index
```

```python
# agg : 여러 함수 적용
# groupby에서 유용함
# *.agg(function)

# Pclass groupby 후 Fare의 max, min
df.groupby('Pclass')['Fare'].agg([max,min])
```

```python
# agg로 칼럼마다 다른 함수 적용
# dictionary로 { 칼럼 : 함수 } 지정

# dictionary 로 Age는 mean, Fare는 max, Sibsp는 sum 지정
func = {'Age' : 'mean', 'Fare' : 'max', 'SibSp' : 'sum'}

# Pclass groupby 후 agg로 각 칼럼별 함수 적용 결과
df.groupby('Pclass').agg(func)
```

## concat

```python
# concat : DataFrame / Series 병합, concatenate
# pd.concat([df1, df2, ..], keys = [ ], join =  , axis =  )
# axis = 0 (default) : 세로로 결합 (행 추가 방향)
# axis = 1  : 가로로 결합 (열 추가 방향)
# keys = []  : 기존 DataFrame별 index 추가
# join = 'outer'/'inner' : 합집합(default) / 교집합


# 병합 대상 DataFrame 생성
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                   'B': ['B0', 'B1', 'B2', 'B3'],
                   'C': ['C0', 'C1', 'C2', 'C3'],
                   'D': ['D0', 'D1', 'D2', 'D3']},
                  index = [0,1,2,3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                   'B': ['B4', 'B5', 'B6', 'B7'],
                   'C': ['C4', 'C5', 'C6', 'C7'],
                   'D': ['D4', 'D5', 'D6', 'D7']},
                  index = [4,5,6,7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                   'B': ['B8', 'B9', 'B10', 'B11'],
                   'C': ['C8', 'C9', 'C10', 'C11'],
                   'D': ['D8', 'D9', 'D10', 'D11']},
                  index = [8,9,10,11])

df4 = pd.DataFrame({'B':['B2','B3','B6','B7'],
                  'D':['D2','D3','D6','D7'],
                  'F':['F2','F3','F6','F7']},
                   index = [2,3,6,7])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
df4
```

```python
# df1, df2, df3 세로로 결합
# axis = 0, 행 추가 방향

con1 = pd.concat([df1, df2, df3])
con1
```

```python
# df1, df2, df3 세로로 결합
# 기존 DataFrame 끼리 구별되도록 keys 지정 -> multi-index

con2 = pd.concat([df1, df2, df3], keys = ['one', 'two', 'three'])
con2
```

```python
# df1, df4 가로로 결합
# 합집합 default
# axis = 1, 열 추가 방향

con3 = pd.concat([df1, df4], axis =1)
con3   # df1, df4의 index 달라서 결측치 발생
```

```python
# df1, df4 가로로 결합, 교집합으로
# join = 'inner'

con4 = pd.concat([df1, df4], join = 'inner', axis = 1)
con4  # index 교집합에 대해서만
```

## merge

```python
# merge : 두 DataFrame 병합 by 공통 칼럼
# pd.merge(df1, df2, how = '', on = '',...)
# how =  'inner'/'outer'/'left'/'right'  : 교집합 default
# on = ''  : how 대상이 되는 열. 
# on 옵션 없으면 df1, df2의 공통 칼럼들 기준으로 merge

# 병합 대상 DataFrame 생성
G = pd.DataFrame({'key': ['K0','K4','K2','K3'],
                    'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']})
H = pd.DataFrame({'key': ['K0','K1','K2','K3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
G
```

```python
H
```

```python
# G, H merge - default 값으로
# 공통칼럼 key에 대해 교집합으로

pd.merge(G,H)
```

```python
# G, H merge - 합집합
# how = 'outer'
# 공통칼럼 key에 대해 합집합으로
# 결측치 생성

pd.merge(G,H, how='outer')
```
